<h3>Retrive Data - SoQL</h3>

Retrived the data using “Socrata Query Language” or “SoQL” using Socarata API

```python
#Import libraries
import numpy as np
import pandas as pd

import plotly.offline as py
import plotly.graph_objs as go
from plotly import tools
import plotly.express as px

#data using SqoQL
soql_health = ('https://data.cityofnewyork.us/resource/uvpi-gqnh.json?' +\
        '$select=health,boroname,spc_common,count(tree_id) as count' +\
        '&$group=health,boroname,spc_common').replace(' ', '%20')

soql_health = pd.read_json(soql_health)
soql_health = soql_health.dropna()
print(soql_health.head())
```

```python
soql_health_steward = ('https://data.cityofnewyork.us/resource/nwxe-4ae8.json?' +\
        '$select=health,boroname,steward,spc_common,count(tree_id) as count' +\
        '&$group=health,boroname,steward,spc_common').replace(' ', '%20')

soql_health_steward = pd.read_json(soql_health_steward)
soql_health_steward = soql_health_steward.dropna()
print(soql_health_steward.head())
```

<h3>Insights</h3>

1. What proportion of trees are in good, fair, or poor health according to the ‘health’ variable?

```python
fig = px.bar(soql_health, x="spc_common", y="count",color="health")
fig.show()
```

From the plot above, Most of the tress are in good health condition

2. Are stewards (steward activity measured by the ‘steward’ variable) having an impact on the health of trees?

```python
fig = px.bar(soql_health_steward, x="spc_common", y="count",color="steward")
fig.show()
```

Steward may not play good role, because pikcing the example of tree 'London planetree', the health of tree is good eventhough proprotion of steward is less, also some tree may not need more attention from steward to keep it in good health
